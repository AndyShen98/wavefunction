```python
from scipy import *
from wavefunction import *
import math
```

```python
h = 6.626e-34
h_ = h/(2*pi)
e = 1.602e-19
Phi0 = h / (2 * e)
cf = 1

ceta = 0.05
J = 0.9725
Ic = 13.3e-6
Ib = J * Ic
Cj = 4.3e-12

mm = Cj * (1 + ceta) * (Phi0/(2*pi))**2 * cf
Ej = Phi0/(2*pi) * Ic / cf

U_current_biased_param = array([Ej, Ic, Ib])
S_param = array([h_, mm])

xmin = -pi
xmax =  pi
N = 750 
```

## Potential

```python
D = arange(xmin, xmax-0.0001, (xmax-xmin)/N )
U = U_current_biased(D, U_current_biased_param);

dmin = math.asin(Ib/Ic)
print "Found dmin = ", dmin
dmax = -math.asin(Ib/Ic) + pi
print "Found dmax = ", dmax

Umin = U_current_biased(dmin, U_current_biased_param);
Umax = U_current_biased(dmax, U_current_biased_param);
dU = Umax - Umin;
delta = dmax-dmin
print "dU = ", dU
```

```python
fig, ax = subplots()

ax.plot(D, U/Ej)
ax.plot(dmin, U_current_biased(dmin, U_current_biased_param) / Ej, 'o')

ax.set_ylim(-1.55, -1.5)
ax.set_xlim(0.8,2.2)
ax.set_xlabel(r'$x$', fontsize=18)
ax.set_ylabel(r'$U(x)/E_J$', fontsize=18);
```

## Calculate the eigenfunctions

```python
M = schrodinger_matrix(xmin, xmax, N, S_param, U_current_biased, U_current_biased_param)

evals, evecs = eigenvectors_sorted(M)
```

## Identify bound states

```python
nbound = 0
boundidx = []

for i in range(0,N):
	if evals[i] > Umin-0.5*dU and evals[i] < Umax+0.5*dU:		
		if inner(evecs[:,i] * (dmin-delta < D) * (D < dmax), evecs[:,i]) > 0.85:
			nbound = nbound + 1
			boundidx.append(i)

print "Found bound states: ", boundidx
```

```python
NN = len(boundidx)

fig, axes = subplots(NN, 1, figsize=(10,NN*1), sharex=True, sharey=True)

for n, m in enumerate(boundidx):
    X = D
    Y = evecs[:,m]
    axes[n].plot(X, Y)

axes[n].set_xlim(0.5, 2)
    
axes[n].set_xlabel(r'$x$', fontsize=18)
#axes[n].set_ylabel(r'$\Psi_n(x)$', fontsize=18);
```

```python
fig, ax = subplots(figsize=(12,8))

ax.plot(D, U/Ej, 'k')
ax.plot(dmin, U_current_biased(dmin, U_current_biased_param) / Ej, '.')

for n, m in enumerate(boundidx):
    X = D
    Y0 = evals[m]/Ej * ones(shape(X))
    Y = 0.001 * evecs[:,m] + Y0

    ax.plot(X, Y0, 'k--')
    ax.plot(X, Y)
    
ax.set_ylim(-1.535, -1.52)
ax.set_xlim(0.8,2.2)
ax.set_xlabel(r'$x$', fontsize=18)
ax.set_ylabel(r'$U(x)$', fontsize=18);
```

## Evaluate transition matrix elements

```python
inner_prod = zeros((len(boundidx), len(boundidx))).astype(float)
expect_pos = zeros((len(boundidx), len(boundidx))).astype(float)
expect_kin = zeros((len(boundidx), len(boundidx))).astype(float)

for i in range(0, len(boundidx)):
	for j in range(0, len(boundidx)):
		l = boundidx[i]
		k = boundidx[j]

		inner_prod[i,j] = inner(evecs[:,k], evecs[:,l])
		expect_pos[i,j] = inner(evecs[:,k], evecs[:,l] * D)
		expect_kin[i,j] = inner(evecs[:,k], derivative(evecs[:,l], D))
```

```python
print "Bound energy-levels =", (evals[boundidx] - evals[boundidx[0]] ) / (h_ * 2*pi * 1e9)
print "Bound energy-levels =", (evals[boundidx] - evals[boundidx[0]] ) / (evals[boundidx[1]]-evals[boundidx[0]])
```
