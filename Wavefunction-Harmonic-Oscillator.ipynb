## Problem parameters

```python
h = 6.626e-34
h = 1
h_ = h/(2*pi)
e = 1.602e-19
cf = 1	      # if cf = h_, use units where h_ = 1

mm = 1          # oscillator mass
omega = 2 * pi  # oscillator frequency in GHz
x0 = 0          # shift in oscillator potiential minimum

U_args = {'w': omega, 'm': mm, 'x0': x0}

S_param = array([h_, mm])

xmin = -2.0 
xmax =  2.0 
N = 750 
```

## Potential

```python
def U_ho(x, args):
    """
    Harmonic oscillator potential
    """
    
    omega = args['w']
    m     = args['m']
    x0    = args['x0']
    
    u = 1/2.0 * m * (omega ** 2) * ((x - x0) ** 2)

    return u
```

### Find the minimum point for the potential

```python
D = arange(xmin, xmax, (xmax-xmin)/N)
```

```python
U = U_ho(D, U_args);

dmin = optimize.fmin(U_ho, [0.0], [U_args])
```

```python
print "Found potential minima at = ", dmin
```

### Plot the potential

```python
fig, ax = subplots()

ax.plot(D, U)
ax.plot(dmin, U_ho(dmin, U_ho_param), 'o')

ax.set_ylim(-10, 80)
ax.set_xlabel(r'$x$', fontsize=18)
ax.set_ylabel(r'$U(x)$', fontsize=18);
```

## Eigenstates and eigenvalues

```python
NN = 10
```

```python
# Calculate the eigenfunctions
M = schrodinger_matrix(xmin, xmax, N, S_param, U_ho, U_args)
evals,evecs = eigenvectors_sorted(M)
```

```python
fig, axes = subplots(NN, 1, figsize=(10,NN*1), sharex=True, sharey=True)

for n in range(NN):
    X = D
    Y = evecs[:,NN-n-1]
    axes[n].plot(X, Y)

axes[n].set_xlim(-1, 1)
    
axes[n].set_xlabel(r'$x$', fontsize=18)
#axes[n].set_ylabel(r'$\Psi_n(x)$', fontsize=18);
```

```python
fig, ax = subplots(figsize=(12,8))

ax.plot(D, U, 'k')
for n in range(10):
    X = D
    Y = evals[n] + evecs[:,n]

    mask = where(Y > U)    
    ax.plot(X[mask], evals[n] * ones(shape(X))[mask], 'k--')

    mask = where(Y > U-2.0)
    ax.plot(X[mask], Y[mask])
    
ax.set_xlim(-1, 1)
ax.set_ylim(0, 10)
ax.set_xlabel(r'$x$', fontsize=18)
ax.set_ylabel(r'$U(x)$', fontsize=18);
```

## Identify bound states

```python
nbound = 0
boundidx = []

for i in range(0,N):
	if nbound < 10: 
		if inner(evecs[:,i], evecs[:,i]) > 0.85:
			nbound = nbound + 1
			boundidx.append(i)
            
print "Found bound states: ", boundidx
```

## Evaluate transition matrix elements

```python
data_vec = [[D]]
for i in boundidx:
	data_vec.append([evecs[:,i]])

inner_prod = zeros((len(boundidx), len(boundidx))).astype(float)
expect_pos = zeros((len(boundidx), len(boundidx))).astype(float)
expect_kin = zeros((len(boundidx), len(boundidx))).astype(float)

for i in range(0, len(boundidx)):
	for j in range(0, len(boundidx)):
		l = boundidx[i]
		k = boundidx[j]

		inner_prod[i,j] = inner(evecs[:,k], evecs[:,l])
		expect_pos[i,j] = inner(evecs[:,k], evecs[:,l] * D)
		expect_kin[i,j] = inner(evecs[:,k], derivative(evecs[:,l], D))
```
