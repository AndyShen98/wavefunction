```python
def jjpq_flux_biased_me(ceta, Cj, cf_p, phi):

	h = 6.626e-34;
	h_ = h/(2*pi);
	e = 1.602e-19;
	Phi0 = h / (2 * e);
	cf = h_;

	Ic = 1.7e-6
	#Cj = 700e-15;
	L = 0.72e-9;

	mm = Cj * (1 + ceta) * (Phi0/(2*pi))**2 * cf;
	beta = 2*pi*L*Ic / Phi0;
	Ej = Phi0/(2*pi) * Ic / cf;

	gamma = phi / beta;

	U_flux_biased_param = array([Ej, beta, gamma])

	S_param = array([h_, mm])

	xmin = -0.7*pi;
	xmax =  4*pi;
	N = 750;


	# -------------------------------------------------------------------------
	# Calculate the eigenfunctions
	#

	M = schrodinger_matrix(xmin, xmax, N, S_param, U_flux_biased, U_flux_biased_param)
	evals,evec = eigenvectors_sorted(M);


	# -------------------------------------------------------------------------
	# Find bound states
	#
	D = arange(xmin, xmax, (xmax-xmin)/N)

	dmin = optimize.fmin(U_flux_biased, [0.5], [U_flux_biased_param])
	print "Found dmin = ", dmin
	dmax = optimize.fmin(U_flux_biased, [2.5], [array([-U_flux_biased_param[0], U_flux_biased_param[1], U_flux_biased_param[2]])])
	print "Found dmax = ", dmax
	delta = dmax-dmin

	Umin = U_flux_biased(dmin, U_flux_biased_param);
	Umax = U_flux_biased(dmax, U_flux_biased_param);
	dU = Umax - Umin;

	nbound = 0;
	boundidx = []

	for i in range(0,N):
        	if evals[i] > Umin-0.5*dU and evals[i] < Umax+0.5*dU:
                	if inner(evec[:,i] * (dmin-delta < D) * (D < dmax), evec[:,i]) > 0.85:
                        	nbound = nbound + 1
                        	boundidx.append(i)

	# -------------------------------------------------------------------------
	# Evaluate matrix elements:
	#
	print "Found bound states: ", boundidx

	inner_prod = zeros((len(boundidx), len(boundidx))).astype(float)
	expect_pos = zeros((len(boundidx), len(boundidx))).astype(float)
	expect_kin = zeros((len(boundidx), len(boundidx))).astype(float)
	energy_levels = zeros(len(boundidx)).astype(float)

	for i in range(0, len(boundidx)):
        	l = boundidx[i]
		energy_levels[i] = evals[l]

        	for j in range(0, len(boundidx)):
        	        k = boundidx[j]

        	        inner_prod[i,j] = inner(evec[:,k], evec[:,l])
        	        expect_pos[i,j] = inner(evec[:,k], evec[:,l] * D)
        	        expect_kin[i,j] = inner(evec[:,k], derivative(evec[:,l], D))

	return energy_levels, expect_pos, expect_kin
```
