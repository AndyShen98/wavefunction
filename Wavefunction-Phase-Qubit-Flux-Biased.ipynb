```python
from scipy import *
from scipy import optimize
from potential import *
from schrodinger_matrix import *
from misc import *
from nummath import *
import math
```

```python
h = 6.626e-34
h_ = h/(2*pi)
e = 1.602e-19
Phi0 = h / (2 * e)
cf = h_

Ic = 1.7e-6
Cj = 700e-15
L = 0.72e-9

mm = Cj * (Phi0/(2*pi))**2 * cf
beta = 2*pi*L*Ic / Phi0
Ej = Phi0/(2*pi) * Ic / cf

phi = 5.0;
gamma = phi / beta;

U_flux_biased_param = array([Ej, beta, gamma])
S_param = array([h_, mm])

xmin = -0.7*pi
xmax =  4*pi
N = 750 
```

## Potential

```python
D = arange(xmin, xmax, (xmax-xmin)/N )
U = U_flux_biased(D, U_flux_biased_param);

dmin = optimize.fmin(U_flux_biased, [0.5], [U_flux_biased_param])
print "Found dmin = ", dmin
dmax = optimize.fmin(U_flux_biased, [2.5], [array([-U_flux_biased_param[0], U_flux_biased_param[1], U_flux_biased_param[2]])])
print "Found dmax = ", dmax

Umin = U_flux_biased(dmin, U_flux_biased_param);
Umax = U_flux_biased(dmax, U_flux_biased_param);
dU = Umax - Umin;
delta = dmax-dmin
print "dU = ", dU
```

```python
fig, ax = subplots()

ax.plot(D, U/Ej)
ax.plot(dmin, U_flux_biased(dmin, U_flux_biased_param) / Ej, 'o')

#ax.set_ylim(-1.55, -1.5)
#ax.set_xlim(0.8,2.2)
ax.set_xlabel(r'$x$', fontsize=18)
ax.set_ylabel(r'$U(x)/E_J$', fontsize=18);
```

```python
fig, ax = subplots()

ax.plot(D, U/Ej)
ax.plot(dmin, U_flux_biased(dmin, U_flux_biased_param) / Ej, 'o')

ax.set_ylim(-2, -1.6)
ax.set_xlim(0.5,3.0)
ax.set_xlabel(r'$x$', fontsize=18)
ax.set_ylabel(r'$U(x)/E_J$', fontsize=18);
```

## Calculate the eigenfunctions

```python
M = schrodinger_matrix(xmin, xmax, N, S_param, U_flux_biased, U_flux_biased_param)

evals,evecs = eigenvectors_sorted(M)
```

## Identify bound states

```python
nbound = 0
boundidx = []

for i in range(0,N):
	if evals[i] > Umin-0.5*dU and evals[i] < Umax+0.5*dU:		
		if inner(evecs[:,i] * (dmin-delta < D) * (D < dmax), evecs[:,i]) > 0.95:
			nbound = nbound + 1
			boundidx.append(i)

print "Found bound states: ", boundidx
```

```python
NN = len(boundidx)

fig, axes = subplots(NN, 1, figsize=(10,NN*1), sharex=True, sharey=True)

for n, m in enumerate(boundidx):
    X = D
    Y = evecs[:,m]
    axes[n].plot(X, Y)

axes[n].set_xlim(0.5, 2.5)
axes[n].set_xlabel(r'$x$', fontsize=18);
```

```python
fig, ax = subplots(figsize=(12,8))

ax.plot(D, U/Ej, 'k')
ax.plot(dmin, U_flux_biased(dmin, U_flux_biased_param) / Ej, '.')

for n, m in enumerate(boundidx):
    X = D
    Y0 = evals[m]/Ej * ones(shape(X))
    Y = 0.01 * evecs[:,m] + Y0

    ax.plot(X, Y0, 'k--')
    ax.plot(X, Y)
    
ax.set_ylim(-1.8, -1.7)
ax.set_xlim(0.5,3.0)
ax.set_xlabel(r'$x$', fontsize=18)
ax.set_ylabel(r'$U(x)$', fontsize=18);
```

## Evaluate transition matrix elements

```python
inner_prod = zeros((len(boundidx), len(boundidx))).astype(float)
expect_pos = zeros((len(boundidx), len(boundidx))).astype(float)
expect_kin = zeros((len(boundidx), len(boundidx))).astype(float)

for i in range(0, len(boundidx)):
	for j in range(0, len(boundidx)):
		l = boundidx[i]
		k = boundidx[j]

		inner_prod[i,j] = inner(evecs[:,k], evecs[:,l])
		expect_pos[i,j] = inner(evecs[:,k], evecs[:,l] * D)
		expect_kin[i,j] = inner(evecs[:,k], derivative(evecs[:,l], D))
```

```python
print "Bound energy-levels =", (evals[boundidx] - evals[boundidx[0]] ) / (h_ * 2*pi * 1e9)
print "Bound energy-levels =", (evals[boundidx] - evals[boundidx[0]] ) / (evals[boundidx[1]]-evals[boundidx[0]])
```
